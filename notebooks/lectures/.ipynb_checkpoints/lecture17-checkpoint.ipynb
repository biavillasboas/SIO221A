#### Transfer function:

We discussed the fact that coherence is analogous to a correlation coefficient.
It tells us if two things vary in tandem in a consistent way, but it doesn't
tell us how big they are or how to use one variable to approximate a second
variable.  If we want to look at relative sizes, or if we want to approximate
a variable $y$ based on its relationship with $x$, in the time domain we think
about finding a regression coefficient.
We're used to solving a least-squares fitting equation of the form

\begin{equation}
{\bf y}=\bf{A}{\bf x}
\end{equation}

with a solution of the form

\begin{equation}
{\bf x} = {(\bf A}^T{\bf A})^{-1} {\bf A}^T {\bf y}
\end{equation}

For illustration we can simplify this to a case where the matrix ${\bf A}$ has
only column, that is where we regress ${\bf y'}$ (with the prime telling us that
the mean has been removed, since we don't want to complicate our least-squares fit)
against just one variable, making
${\bf A}$ a column vector (e.g. $\bf z'$).  In this case $x$ becomes a scalar and the matrix
inverse $({\bf A}^T{\bf A})^{-1}$ is just the reciprocal of the variance of $\bf z'$.

\begin{equation}
x = \frac{\langle y'z'\rangle}{\langle z'^2\rangle}.
\end{equation}

Compare this with the correlation coefficient of the demeaned variables

\begin{equation}
r =  \frac{\langle y'z'\rangle}{\sqrt{\langle y'^2\rangle \langle z'^2\rangle}}.
\end{equation}

Viewed in this way, the correlation coefficient $r$ and the regression coefficient $x$
(another term for the least-squares fit with only one variable) look nearly
the same, aside from the normalization.
Both the correlation coefficient and the regression coefficient convey
useful information.  And that might make you think that
in the Fourier-transform domain
there should be a form analogous to regression.

The term for the Fourier domain analog to regression is the transfer function:

\begin{equation}
\hat{H}_{xy}(f) = \frac{\hat{G}_{xy}(f)}{\hat{G}_{xx}(f)},
\end{equation}

which provides a (complex-numbered) recipe for mapping from $x$ to $y$.

Formally, we talk about the transfer function when we think about
constructing a linear system:

\begin{equation}
\mathcal{L}(y(t)) = x(t)
\end{equation}

If $\mathcal{L}$ is a linear operator, then we could think of this relationship
as a convolution:

\begin{equation}
y_t = \int_{-\infty}^{\infty} h(u) x(t-u)\, du
\end{equation}

or if we Fourier transform, this would state:

\begin{equation}
Y(f) = H(f) X(f).
\end{equation}

#### Transfer functions (or gain functions):  a proper example

If we want to look at relative sizes of two Fourier-transformed quantities,
we can look at the
transfer function (also known as the gain function):

\begin{equation}
\hat{H}_{xy}(f) = \frac{\hat{G}_{xy}(f)}{\hat{G}_{xx}(f)},
\end{equation}

which provides a (complex-numbered) recipe for mapping from $x$ to $y$.
Then, if

\begin{equation}
y_t = \int_{-\infty}^{\infty} h(u) x(t-u)\, du,
\end{equation}

the Fourier transform is:

\begin{equation}
Y(f) = H(f) X(f).
\end{equation}

Consider it this way.  Suppose

\begin{equation}
x(t) = \frac{d^2y}{dt^2} + \alpha \frac{dy}{dt} + \beta y
\end{equation}

Then by Fourier transforming, we have:

\begin{eqnarray}
X(f) & = & -f^2 Y(f)+i\alpha f Y(f) + \beta Y(f) \\
 & = &  Y(f)\left[\beta -f^2 +i\alpha f\right]
\end{eqnarray}

so

\begin{equation}
Y(f) = \frac{1}{\left[\beta -f^2 +i\alpha f\right]} X(f)
\end{equation}

and

\begin{equation}
H(f) = \frac{1}{\left[\beta -f^2 +i\alpha f\right]}
\end{equation}


This is a nice framework for solving differential equations, but can we
use it to gain insights into our data as well?
First some rules:

1. Linearity:  If a given linear system has an input $x_1(t)$ which corresponds to an output $y_1(t)$, and input $x_2(t)$ corresponds to output $y_2(t)$m then a summed input $x(t)=\alpha x_1(t)+\beta x_2(t)$, will produce an output $y(t)=\alpha y_1(t)+\beta y_2(t)$.

2. Time invariance:  If an input is delayed in time by $\tau$, then the output is as well:  If $x(t) \rightarrow x(t+\tau)$, then $y(t) \rightarrow y(t+\tau)$.

3. Causality:  If h(t) represents an impulse, then it should be zero for $t<0$.  A response cannot occur before the forcing.

4. Sequential application:  If the output of one linear system is an input to a second system, then the frequency response is

\begin{equation}
H_{12}(f) = H_1(f)\cdot H_2(f)
\end{equation}

So suppose we measure $y(t)$ and $x(t)$.  Can we determine $h$ or $H$?
We know that

\begin{equation}
Y(f) = H(f)X(f)
\end{equation}

Let's multiply both sides of the equation by the complex conjugate of $X$ to
form the cross-spectrum:

\begin{equation}
\frac{Y(f)X^*(f)}{T} = H(f)\frac{X(f)X^*(f)}{T}
\end{equation}

This becomes

\begin{equation}
G_{xy}(f)= H(f)G_{xx}(f)
\end{equation}

so

\begin{equation}
 H(f)=\frac{G_{xy}(f)}{G_{xx}(f)}
\end{equation}

#### Salinity spiking examples

A classic example comes from the "salinity spiking" problem, which arises
due to the different response times of temperature and
conductivity sensors:

\begin{equation}
S=S(T(t),C(t-\Delta t)).
\end{equation}

Our challenge now is to use the transfer (or gain) function to assess the
salinity spiking, and
maybe even to correct it.
Conductivity is strongly dependent on temperature, and we have
to remove the temperature effect to determine salinity.  But the
faster response times of conductivity
sensors relative to temperature sensors are a source of confusion.
Here's a basic procedure.

1. Identify a segment of the water column in which temperature and conductivity should be well behaved, with fluctuations due to temperature only.  (This isn't essential, but it will give us a good shot at unraveling the sensor response time issues that lead to salinity or density spiking.) We'll identify the true values as $T$ and $C$, and the measured values as $\hat{T}$ and $\hat{C}$.

2. Collect a lot of profiles of data. 

3. Now treat this as a linear system:

\begin{eqnarray}
 \hat{T}(k) & = & H_T(k) T(k)\\
 \hat{C}(k) & = & H_C(k) T(k)\\
\end{eqnarray}

Here $H_T(k)$ is the spatial/frequency response of the temperature sensor,
and $H_C(k)$ is the spatial/frequency response of the conductivity sensor.
The use of $T(k)$ in the conductivity equation might seem a little crazy,
but it's really important, since we're asserting that salinity is unimportant
in our (hypothetical) study region.

4. Compute cross spectra:

\begin{eqnarray}
 \hat{G}_{TT}(k) & = & \frac{\langle \hat{T}^*(k)\hat{T}(k)\rangle}{N\Delta t}\\
 \hat{G}_{CC}(k) & = & \frac{\langle \hat{C}^*(k)\hat{C}(k)\rangle}{N\Delta t}\\
 \hat{G}_{TC}(k) & = & \frac{\langle \hat{T}^*(k)\hat{C}(k)\rangle}{N\Delta t}\\
\end{eqnarray}

Then if we substitute in the expessions linking the observed values to the
true values we obtain:

\begin{eqnarray}
\frac{\langle\hat{T}^*(k)\hat{C}(k)\rangle}{N\Delta t} & = & \frac{\langle(H_T(k)T(k))^* H_C(k)T(k)\rangle}{N\Delta t} \\
 & = & [H_T^*(k)H_C(k)] \frac{\langle T^*(k) T(k)\rangle}{N\Delta t} \\
\hat{G}_{TC}(k) & = &  [H_T^*(k)H_C(k)] G_{TT}(k)
\end{eqnarray}

The same applies for the temperature spectrum:

\begin{eqnarray}
 \hat{G}_{TT}(k) & = & \frac{\langle \hat{T}^*(k)\hat{T}(k)\rangle}{N\Delta t}\\
 & = & [H_T^*(k)H_T(k)] G_{TT}(k)
\end{eqnarray}

So the ratio of these becomes:

\begin{equation}
\frac{\hat{G}_{TC}(k)}{\hat{G}_{TT}(k)} = \frac{H_T^*(k)H_C(k)}{|H_T|^2} =
\frac{H_C(k)}{H_T(k)}
\end{equation}

This means that even without knowing the response function $H$,
we can compute the ratio of the response functions
from the transfer function, the ratio of the  cross-spectrum to the
spectrum.

An analogous relationship also holds:

\begin{equation}
\frac{\hat{G}_{TC}(k)}{\hat{G}_{CC}(k)} = \frac{H_T^*(k)H_C(k)}{|H_C|^2} =
\frac{H_T(k)}{H_C(k)}
\end{equation}

5. Now use this information to correct the conductivity sensor to have the same response as the temperature sensor.  Here we'll define our corrected conductivity as $\hat{\hat{C}}(k)$, and we want to understand its relationship with the observed temperature $\hat{T}(k)$ and the true temperature $T(k)$.

\begin{equation}
\hat{\hat{C}}(k) = \alpha\hat{T}(k) = \alpha H_T(k) T(k)
\end{equation}

This means we need a correction of the form:

\begin{equation}
\hat{\hat{C}}(k) = \hat{C}(k)\cdot P(k) = \alpha\hat{T}(k),
\end{equation}

and our task is to figure out $P(k)$.
We can also write:

\begin{equation}
\hat{C}(k)\cdot P(k) = H_C(k)T(k)P(k)
\end{equation}

so putting this together:

\begin{equation}
\alpha H_T(k) T(k)  = H_C(k)T(k)P(k)
\end{equation}

Thus

\begin{equation}
P(k) = \frac{\alpha H_T(k)}{H_C(k)} = \alpha \frac{\hat{G}_{TC}(k)}{\hat{G}_{CC}(k)}
\end{equation}

where $\alpha$ is real.
So we do a bit of curve fitting to optimize our correction.


A typical correction might allow for errors both in the response time and a
direct time lag:

\begin{equation}
\tau \frac{d\hat{T}(t)}{dt} + \hat{T}(t) = T(t-L)
\end{equation}

(from Giles and McDougall, Deep-Sea Research, 1986)
and this suggests corrections both in the frequency and time domain, either by
minimizing phase differences or by maximizing correlation.
We can Fourier transform this to find:

\begin{equation}
-i2\pi f\tau {\cal{F}}(\hat{T}) + {\cal{F}}(\hat{T}) = {\cal{F}}(T)e^{-i2\pi fL},
\end{equation}

implying that

\begin{equation}
(1-i2\pi f\tau)e^{i2\pi f L} {\cal{F}}(\hat{T}) = {\cal{F}}(T).
\end{equation}

Since conductivity has the fast response, one strategy is to treat $\hat{C}$
as behaving like the true temperature $T$.  So hypothetically:

\begin{eqnarray}
 \hat{T}(k) & = & H_T(k) T(k) = \frac{1}{(1-i2\pi f\tau)e^{i2\pi f L}} T(k)\\
 \hat{C}(k) & = & H_C(k) T(k) = T(k)\\
\end{eqnarray}

implying a correction P(k) of the form:

\begin{equation}
P(k) = \frac{\alpha H_T(k)}{H_C(k)} = \alpha \frac{1}{(1-i2\pi f\tau)e^{i2\pi f L}}.
\end{equation}

#### Transfer function: Practical implication

You might be wondering exactly how to compute a transfer function, so here's a practical example.

```python
import sys
sys.path.append('../../tools/')

import numpy as np
from numpy import pi, cos, sin
import matplotlib.pyplot as plt

from segment import overlapping_segments1D 

np.random.seed(0)

lamb = 10 # 10 m wavelength
V = 0.3 # 0.3 m/s propagation
n2s = 0.2 # noise-to-signal ratio

N = 5000
time = np.arange(1, N+1)
x = n2s*np.random.randn(N) + cos(2*pi/lamb*V*time)
y = n2s*np.random.randn(N) + cos(2*pi/lamb*V*(time)+pi/2)
```

```python
# start with a segmented data approach:
p = N//10
xx = overlapping_segments1D(x, segment_length=p, overlap=0.5)
yy = overlapping_segments1D(y, segment_length=p, overlap=0.5)
M = xx.shape[0]
```

```python
win = np.hanning(p)*np.ones([M, p])*np.sqrt(8/3)
fxx = np.fft.fft(xx*win, axis=1)
fyy = np.fft.fft(yy*win, axis=1)

sx = np.sum(abs(fxx[:, :p//2+1]/p)**2, axis=0)
sx[1:-1] = 2*sx[1:-1]

sy = np.sum(abs(fyy[:, :p//2+1]/p)**2, axis=0)
sy[1:-1] = 2*sy[1:-1]

cxy = np.sum(fxx[:, :p//2+1].conjugate()*fyy[:, :p//2+1]/p/p, axis=0) 
cxy[1:-1] = 2*cxy[1:-1]
C = abs(cxy)**2/(sx*sy)

phase_C = np.arctan2(-cxy.imag, cxy.real)

transfer = cxy/sx
phase = np.arctan2(-transfer.imag, transfer.real)
```

```python
plt.figure(figsize=(10, 8))
plt.semilogx(np.arange(p//2+1), C, label='coherence' )
plt.semilogx(np.arange(p//2+1), abs(transfer), label='transfer function amplitude')
plt.xlabel('frequency', fontsize=14)
plt.ylabel('coherence or transfer function',fontsize=14)
plt.legend(loc=2, fontsize=14)
```

*Figure 1: Transfer function example (red) compared with coherence (blue) for an artificial data sets consisting of white noise plus either a sine or cosine at one frequency.*

In this case, you'll see that the transfer function peak happens to be about the same size
as the coherence, while the background noise for the transfer function is significantly
larger than the background noise for the coherence.   These results are entirely a function
of the data that we use.  The high background noise level stems from the fact that the
coherence is scaled by $x$, which can be very small.

#### Noise and coherence

Finally, let's take a quick look at
what noise does to coherence and to the transfer function.
Suppose we start with two related signals:

\begin{equation}
y(t) = c_1 x(t) + n(t),
\end{equation}

where $n(t)$ is noise.  Can we determine $c_1$ and $n(t)$ using the
cross-spectrum?   Formally, the cross spectrum is:

\begin{eqnarray}
S_{XY}(f) & = &\frac{\langle X^*(f)Y(f)\rangle}{T}\\
 & = & \frac{c_1  {\langle X^*(f)X(f)\rangle + \langle X^*(f)N(f)\rangle}}{T}\\
 & = & c_1 S_{XX}
\end{eqnarray}

The squared coherence is of course:

\begin{equation}
\text{Coh}(f)^2 = \gamma_{XY}^2 = \frac{|S_{XY}|^2}{S_{XX}S_{YY}}
\end{equation}

So what is $S_{YY}$?
We can compute the spectrum of $y$:

\begin{eqnarray}
S_{YY} & = & \frac{\left[ c_1^2 \langle X^*X\rangle + c_1(\langle X^*N\rangle + \langle XN^*\rangle
 + \langle N^*N\rangle\right]}{T}\\
 & = & c_1^2 S_{XX} + S_{NN},
\end{eqnarray}

where $X$ and $N$ are assumed to be uncorrelated.
Then the squared coherence is:

\begin{eqnarray}
\gamma_{XY}^2(f) & = & \frac{c_1^2 S_{XX}(f)^2}{S_{XX}(c_1^2S_{XX} + S_{NN})}\\
& = & \frac{1}{ \left(1 + \frac{S_{NN}(f)}{c_1^2S_{XX}(f)}\right)}
\end{eqnarray}

The final term in the denominator is a measure of the noise-to-signal
ratio.  (In our example, we imposed it from the beginning.)  So if we
knew a lot about the causal relations between our records, we could use
the coherence to extract a measure of the noise-to-signal ratio.

#### Noise and the transfer function

Now let's introduce noise in the Fourier transform domain, which perhaps
implies a slightly more complicated relationship
between $x$ and $y$:

\begin{equation}
Y(f) = X(f)H(f) + N(f)
\end{equation}

If we multiply through by $X^*$:

\begin{equation}
\langle X^*Y\rangle = \langle X^* X\rangle H(f) + \langle X^*N\rangle
\end{equation}

Since the signal is uncorrelated with noise, this still gives us

\begin{equation}
 H(f)=\frac{G_{xy}(f)}{G_{xx}(f)},
\end{equation}

so noise appears to have no impact on the results, but is it all so rosy?

Alternatively, we might imagine that our forcing $x$ is noisy, so that

\begin{equation}
Y(f) = \left[X(f)+N(f)\right] H(f)
\end{equation}

In doing this, we assume that the noise associated with $X$ is uncorrelated
with the signal $Y$---in other words, we assume that the response $y(t)$
should be responding to $x(t)$, but we've mismeasured the forcing as
$x(t)+n(t)$.
Then:

\begin{equation}
\langle (X+N)^*Y\rangle = \langle (X+N)^* (X+N)\rangle H(f)
\end{equation}

And since $y$ and $n$ are uncorrelated in this formulation,

\begin{equation}
 \hat{H}(f)=\frac{G_{xy}(f)}{G_{xx}(f)+G_{nn}(f)},
\end{equation}

which is biased low relative to the true response.  (But the phase is unbiased.)
This is analogous to the
noise-related formulation that we looked at for coherence.

We could test all of this with an example:

```python
x = x + np.random.randn(N) + cos(2*pi/lamb*V*time)

# start with a segmented data approach:
p = N//10
xx = overlapping_segments1D(x, segment_length=p, overlap=0.5)
yy = overlapping_segments1D(y, segment_length=p, overlap=0.5)
M = xx.shape[0]

win = np.hanning(p)*np.ones([M, p])*np.sqrt(8/3)
fxx = np.fft.fft(xx*win, axis=1)
fyy = np.fft.fft(yy*win, axis=1)

sx = np.sum(abs(fxx[:, :p//2+1]/p)**2, axis=0)
sx[1:-1] = 2*sx[1:-1]

sy = np.sum(abs(fyy[:, :p//2+1]/p)**2, axis=0)
sy[1:-1] = 2*sy[1:-1]

cxy = np.sum(fxx[:, :p//2+1].conjugate()*fyy[:, :p//2+1]/p/p, axis=0) 
cxy[1:-1] = 2*cxy[1:-1]

transferx = cxy/sx
transfery = cxy/sy

phasex = np.arctan2(-transferx.imag, transferx.real)
phasey = np.arctan2(-transfery.imag, transfery.real)
```

```python
plt.figure(figsize=(10, 8))
plt.semilogx(np.arange(p//2+1), abs(transfer), label='$H_{xy}$')
plt.semilogx(np.arange(p//2+1), abs(transferx), label='$H_{x+n,y}$')
plt.semilogx(np.arange(p//2+1), abs(transfery), label='$H_{y,x+n}$')
plt.xlabel('frequency', fontsize=14)
plt.ylabel('transfer function',fontsize=14)
plt.legend(loc=2, fontsize=14)
```

*Figure 2: Transfer function for baseline case, vs examples with noise added. Noise substantially modifies the transfer function, and its impact depends on whether it appears in the denominator.*

```python
plt.figure(figsize=(10, 8))
plt.semilogx(np.arange(p//2+1), phase, label='$H_{xy}$')
plt.semilogx(np.arange(p//2+1), phasex, label='$H_{x+n,y}$')
plt.semilogx(np.arange(p//2+1), phasey, label='$H_{y,x+n}$')
plt.xlabel('frequency', fontsize=14)
plt.ylabel('transfer function phase',fontsize=14)
plt.legend(loc=2, fontsize=14)
```

*Figure 3: Phase for the transfer function examples shown in Figure 2. Phase depends on the signal, but does not depend on the spectral normalization that appears in the denominator, so the red and yellow lines are the same.*
