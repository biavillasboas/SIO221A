# A few tips to get started with homework 4

Here I will illustrate how to look for a chunk of data that is regularly spaced.

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import datetime
```

```python
sccoos_thredds = 'http://sccoos.org/thredds/dodsC/autoss/scripps_pier-2019.nc'
ds = xr.open_dataset(sccoos_thredds)
```

```python
ds.pressure.plot()
```

```python
tdelta = np.diff(ds.pressure.time[:])
```

```python
tdelta
```

```python
(tdelta == datetime.timedelta(seconds=240)).all()
```

```python
ind = tdelta == datetime.timedelta(seconds=240)
```

```python
plt.figure(figsize = (12, 6))
plt.plot(ds.pressure.time[:-1].values, ind.astype('int'))
```

```python
plt.figure(figsize = (12, 6))
plt.plot(ds.pressure.time[:-1].values[102826:112938], ind.astype('int')[102826:112938])
```
